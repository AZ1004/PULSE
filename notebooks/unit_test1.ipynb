```python
import torch
import sys
import os

# This adds the parent directory (PULSE root) to the system path
sys.path.append(os.path.abspath(os.path.join('..')))
from losses.nt_xent import NTXentLoss

device = "cuda" if torch.cuda.is_available() else "cpu"
loss_fn = NTXentLoss(temperature=0.5, device=device)

# ── Test 1: identical pairs → loss should be near 0 ──────────────────────────
z = torch.randn(8, 128).to(device)
loss_identical = loss_fn(z, z.clone())
print(f"Identical pairs loss : {loss_identical.item():.4f}  (expect: near 0)")

# ── Test 2: random noise pairs → loss should be high ─────────────────────────
z_i = torch.randn(8, 128).to(device)
z_j = torch.randn(8, 128).to(device)
loss_random = loss_fn(z_i, z_j)
print(f"Random pairs loss    : {loss_random.item():.4f}  (expect: ~log(2N-1) = {torch.log(torch.tensor(15.0)):.4f})")

# ── Test 3: output is a scalar ────────────────────────────────────────────────
assert loss_random.shape == torch.Size([]), "Loss must be a scalar"
print("✅ All NT-Xent tests passed")
```
