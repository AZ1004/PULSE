```python
from pathlib import Path
from torch.utils.data import DataLoader
import sys
import os

# This adds the parent directory (PULSE root) to the system path
sys.path.append(os.path.abspath(os.path.join('..')))

from data.dataset import BUSIDataset, BUSIDatasetSSL
from augmentations.ssl_transforms import SimCLRAugment  # your existing file

root = Path(r"C:\Users\asmaa\Desktop\BUSI\Dataset_BUSI_with_GT")

# ── 1. Supervised dataset ──────────────────────────────────────────────────────
dataset = BUSIDataset(root=root, image_size=224)
print(dataset)
# Should print: Total samples : 780, benign: 437, malignant: 210, normal: 133

img, label = dataset[0]
print(f"Image type : {type(img)}")   # PIL Image (no transform set yet)
print(f"Label      : {label}")       # 0, 1, or 2

# ── 2. SSL dataset ─────────────────────────────────────────────────────────────
ssl_transform = SimCLRAugment(size=224)
ssl_dataset   = BUSIDatasetSSL(busi_dataset=dataset, ssl_transform=ssl_transform)

v1, v2, label = ssl_dataset[0]
print(f"View 1 shape : {v1.shape}")  # torch.Size([3, 224, 224])
print(f"View 2 shape : {v2.shape}")  # torch.Size([3, 224, 224])
print(f"Label        : {label}")

# ── 3. DataLoader smoke test ───────────────────────────────────────────────────
loader = DataLoader(ssl_dataset, batch_size=8, shuffle=True, num_workers=0)
batch_v1, batch_v2, batch_labels = next(iter(loader))
print(f"Batch shape : {batch_v1.shape}")  # torch.Size([8, 3, 224, 224])
print("✅ DataLoader works")
```
